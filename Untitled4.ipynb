```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting histograms for the numerical columns
plt.figure(figsize=(14, 10))

# Income Distribution
plt.subplot(2, 3, 1)
sns.histplot(df['Income'], kde=True, color='skyblue')
plt.title('Income Distribution')

# Spending Score Distribution
plt.subplot(2, 3, 2)
sns.histplot(df['Spending_Score'], kde=True, color='green')
plt.title('Spending Score Distribution')

# Membership Years Distribution
plt.subplot(2, 3, 3)
sns.histplot(df['Membership_Years'], kde=True, color='orange')
plt.title('Membership Years Distribution')

# Purchase Frequency Distribution
plt.subplot(2, 3, 4)
sns.histplot(df['Purchase_Frequency'], kde=True, color='purple')
plt.title('Purchase Frequency Distribution')

# Last Purchase Amount Distribution
plt.subplot(2, 3, 5)
sns.histplot(df['Last_Purchase_Amount'], kde=True, color='red')
plt.title('Last Purchase Amount Distribution')

plt.tight_layout()
plt.show()
```

```python
# Correlation matrix
correlation_matrix = df.corr()

# Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter plots to explore relationships
plt.figure(figsize=(14, 10))

# Income vs. Spending Score
plt.subplot(2, 3, 1)
sns.scatterplot(x='Income', y='Spending_Score', data=df, color='blue')
plt.title('Income vs Spending Score')

# Membership Years vs. Purchase Frequency
plt.subplot(2, 3, 2)
sns.scatterplot(x='Membership_Years', y='Purchase_Frequency', data=df, color='green')
plt.title('Membership Years vs Purchase Frequency')

# Income vs. Purchase Frequency
plt.subplot(2, 3, 3)
sns.scatterplot(x='Income', y='Purchase_Frequency', data=df, color='orange')
plt.title('Income vs Purchase Frequency')

# Last Purchase Amount vs. Spending Score
plt.subplot(2, 3, 4)
sns.scatterplot(x='Last_Purchase_Amount', y='Spending_Score', data=df, color='purple')
plt.title('Last Purchase Amount vs Spending Score')

# Membership Years vs. Spending Score
plt.subplot(2, 3, 5)
sns.scatterplot(x='Membership_Years', y='Spending_Score', data=df, color='red')
plt.title('Membership Years vs Spending Score')

plt.tight_layout()
plt.show()
```

```python
# Outlier detection using boxplots
plt.figure(figsize=(14, 10))

# Income Boxplot
plt.subplot(2, 3, 1)
sns.boxplot(x=df['Income'], color='skyblue')
plt.title('Income Boxplot')

# Spending Score Boxplot
plt.subplot(2, 3, 2)
sns.boxplot(x=df['Spending_Score'], color='green')
plt.title('Spending Score Boxplot')

# Membership Years Boxplot
plt.subplot(2, 3, 3)
sns.boxplot(x=df['Membership_Years'], color='orange')
plt.title('Membership Years Boxplot')

# Purchase Frequency Boxplot
plt.subplot(2, 3, 4)
sns.boxplot(x=df['Purchase_Frequency'], color='purple')
plt.title('Purchase Frequency Boxplot')

# Last Purchase Amount Boxplot
plt.subplot(2, 3, 5)
sns.boxplot(x=df['Last_Purchase_Amount'], color='red')
plt.title('Last Purchase Amount Boxplot')

plt.tight_layout()
plt.show()
```

```python
# Define a new column to segment customers based on 'Last_Purchase_Amount'
df['Purchase_Segment'] = pd.cut(df['Last_Purchase_Amount'], bins=[0, 5000, 10000, 15000],
                                labels=['Low', 'Medium', 'High'])

# Count the number of customers in each segment
purchase_segment_counts = df['Purchase_Segment'].value_counts()

# Plot the distribution of purchase segments
purchase_segment_counts.plot(kind='bar', color='teal', figsize=(8, 6))
plt.title('Customer Segments Based on Last Purchase Amount')
plt.xlabel('Purchase Segment')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Membership Years vs. Spending Score (with a regression line)
plt.figure(figsize=(8, 6))
sns.regplot(x='Membership_Years', y='Spending_Score', data=df, scatter_kws={'color': 'orange'}, line_kws={'color': 'purple'})
plt.title('Membership Years vs Spending Score')
plt.xlabel('Membership Years')
plt.ylabel('Spending Score')
plt.show()
```

```python
# Calculate Customer Lifetime Value (CLV)
df['Average_Purchase_Value'] = df['Last_Purchase_Amount'] / df['Purchase_Frequency']
df['CLV'] = df['Average_Purchase_Value'] * df['Purchase_Frequency'] * df['Membership_Years']

# Add Discount Impact (Assuming there's a discount column or we calculate one from purchase behavior)
# Here we'll assume a 5% discount impact on the `Last_Purchase_Amount` for demonstration.
df['Discount_Amount'] = df['Last_Purchase_Amount'] * 0.05  # Example: 5% discount impact
df['Adjusted_Purchase_Amount'] = df['Last_Purchase_Amount'] + df['Discount_Amount']

# Segment customers into 'High Potential', 'Medium Potential', and 'Low Potential' based on CLV
df['Sales_Potential'] = pd.cut(df['CLV'], bins=[0, 100000, 500000, 1000000], labels=['Low', 'Medium', 'High'])
```

```python
# Plotting the relationship between CLV and Adjusted Purchase Amount
plt.figure(figsize=(8, 6))
sns.scatterplot(x='CLV', y='Adjusted_Purchase_Amount', data=df, hue='Sales_Potential', palette='coolwarm', s=100)
plt.title('CLV vs Adjusted Purchase Amount')
plt.xlabel('Customer Lifetime Value (CLV)')
plt.ylabel('Adjusted Purchase Amount (After Discount)')
plt.show()
```

```python
# Let's assume that customers who received a discount have an `Is_Discounted` column
# In this case, we'll generate a random `Is_Discounted` for demonstration.
df['Is_Discounted'] = np.random.choice([0, 1], size=len(df), p=[0.3, 0.7])

# Plot the impact of discounts on spending
plt.figure(figsize=(8, 6))
sns.boxplot(x='Is_Discounted', y='Last_Purchase_Amount', data=df, palette='coolwarm')
plt.title('Impact of Discounts on Last Purchase Amount')
plt.xlabel('Is Discounted')
plt.ylabel('Last Purchase Amount')
plt.show()
```

```python
# Identify top 10 customers based on CLV (Customer Lifetime Value)
top_10_customers = df.nlargest(10, 'CLV')

# Plot the top 10 customers by CLV
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_customers['Number'], y=top_10_customers['CLV'], palette='Blues_d')
plt.title('Top 10 Customers by CLV')
plt.xlabel('Customer Number')
plt.ylabel('Customer Lifetime Value')
plt.show()
```
